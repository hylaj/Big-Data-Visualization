## Kernel PCA - Experiments
PCA is a linear method. That is it can only be applied to datasets which are linearly separable. It does an excellent job for datasets, which are linearly separable. But, if we use it to non-linear datasets, we might get a result which may not be the optimal dimensionality reduction. Kernel PCA uses a kernel function to project dataset into a higher dimensional feature space, where it is linearly separable. It is similar to the idea of Support Vector Machines.

```python

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 500, noise = 0.02, random_state = 417)

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

Let’s apply PCA on this dataset:

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

PCA failed to distinguish the two classes.



```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(kernel ='rbf', gamma = 15)
X_kpca = kpca.fit_transform(X)

plt.title("Kernel PCA")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c = y)
plt.show()
```

```python
kpca_rbf_low = KernelPCA(kernel='rbf', gamma=0.1)
X_kpca_1 = kpca_rbf_low.fit_transform(X)

plt.scatter(X_kpca_1[:, 0], X_kpca_1[:, 1], c=y, cmap='coolwarm')
plt.title('RBF Kernel (Gamma = 0.1)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()




kpca_rbf_high = KernelPCA(kernel='rbf', gamma=1)
X_kpca_2 = kpca_rbf_high.fit_transform(X)

plt.scatter(X_kpca_2[:, 0], X_kpca_2[:, 1], c=y, cmap='coolwarm')
plt.title('RBF Kernel (Gamma = 1)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

kpca_rbf_high = KernelPCA(kernel='rbf', gamma=5)
X_kpca_2 = kpca_rbf_high.fit_transform(X)

plt.scatter(X_kpca_2[:, 0], X_kpca_2[:, 1], c=y, cmap='coolwarm')
plt.title('RBF Kernel (Gamma = 5)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

kpca_rbf_high = KernelPCA(kernel='rbf', gamma=10)
X_kpca_2 = kpca_rbf_high.fit_transform(X)

plt.scatter(X_kpca_2[:, 0], X_kpca_2[:, 1], c=y, cmap='coolwarm')
plt.title('RBF Kernel (Gamma = 10)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()


kpca_poly = KernelPCA(kernel='poly', degree=2)
X_kpca_3 = kpca_poly.fit_transform(X)

plt.scatter(X_kpca_3[:, 0], X_kpca_3[:, 1], c=y, cmap='coolwarm')
plt.title('Polynomial Kernel (Degree 2)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show() 

kpca_poly = KernelPCA(kernel='poly', degree=3)
X_kpca_3 = kpca_poly.fit_transform(X)

plt.scatter(X_kpca_3[:, 0], X_kpca_3[:, 1], c=y, cmap='coolwarm')
plt.title('Polynomial Kernel (Degree 3)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

kpca_poly = KernelPCA(kernel='poly', degree=5)
X_kpca_3 = kpca_poly.fit_transform(X)

plt.scatter(X_kpca_3[:, 0], X_kpca_3[:, 1], c=y, cmap='coolwarm')
plt.title('Polynomial Kernel (Degree 5)', fontsize=16, pad=15)
plt.xlabel("Kernel PC1", fontsize=12)
plt.ylabel("Kernel PC2", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
